```python
# 数据切分，将原有数据划分为训练集、验证集、测试集
```

```python
import os
import random
#import shutil
from shutil import copy2
```

```python
from google.colab import drive
drive.mount('/content/drive/')
```

```python
#比例
scale = [0.7, 0.3]
```

```python
flower_classes = ['daisy','sunflower','bauhinia','huang wood',
                 'kakop'] 
```

```python
for name in flower_classes:
    datadir_normal = "drive/MyDrive/graduation project/work/enhance data/"+name+"/"  #原文件夹
    # if not os.path.exists(datadir_normal):  #如果不存在这个文件夹，就创造一个
    #     os.makedirs(datadir_normal)
    all_data = os.listdir(datadir_normal)#（图片文件夹）
    num_all_data = len(all_data)
    print(name+ "类图片数量: " + str(num_all_data) )
    index_list = list(range(num_all_data))
    #print(index_list)
    random.shuffle(index_list)
    num = 0
    
    trainDir = "drive/MyDrive/graduation project/work/devide data/train/"+name#（将训练集放在这个文件夹下）
    if not os.path.exists(trainDir):  #如果不存在这个文件夹，就创造一个
        os.makedirs(trainDir)
            
    testDir = "drive/MyDrive/graduation project/work/devide data/test/"+name#（将测试集放在这个文件夹下）        
    if not os.path.exists(testDir):
        os.makedirs(testDir)
            
    for i in index_list:
        fileName = os.path.join(datadir_normal, all_data[i])
        if num < num_all_data*scale[0]:
            #print(str(fileName))
            copy2(fileName, trainDir)
        else:
            copy2(fileName, testDir)
        num += 1
```
