```
import requests
import json
```

```
from google.colab import drive
drive.mount('/content/drive/')
```

```
n = 1
page = 8 # 读取数据的页数，每页30条数据
pn = 30
url = 'https://image.baidu.com/search/acjson?'
h = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.110 Safari/537.36'
}
img_path_list = []
```

```
while n <= page:
  p = {
    'tn': 'resultjson_com', 'logid': '8846269338939606587', 'ipn': 'rj', 'ct': '201326592', 'is': '', 'fp': 'result', 'queryWord': '洋紫荆花', 'cl': '2', 'lm': '-1', 'ie': 'utf-8', 'oe': 'utf-8', 'adpicid': '', 'st': '-1', 'z':'' , 'ic':'' , 'hd': '', 'latest': '', 'copyright': '', 'word': '洋紫荆花', 's':'' , 'se':'' , 'tab': '', 'width': '', 'height': '', 'face': '0', 'istype': '2', 'qc': '', 'nc': '1', 'fr': '', 'expermode': '', 'force': '', 'cg': 'girl', 'pn': pn, 'rn': '30', 'gsm': '1e'
  }
  response = requests.get(url = url,headers = h,params = p).text
  res = json.loads(response)['data']
  del res[-1]
  for r in res:
    img_path_list.append(r['thumbURL'])
  n += 1
  pn += 30
```

```
idx = 1
for u in img_path_list:
  img = requests.get(url = u).content
  img_path ='drive/MyDrive/graduation project/work/data/bauhinia/image_' + str(idx) + '.jpg'
  with open(img_path,'wb') as f:
   f.write(img) 
  idx += 1 
```
