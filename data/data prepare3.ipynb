```python
import os
from PIL import Image
```

```python
from google.colab import drive
drive.mount('/content/drive/')
```

```python
flowe_classes = ['daisy','sunflower','bauhinia','huang wood',
                 'kakop'] 
```

```python
# 水平翻转
for name in flowe_classes:
    a = os.listdir('drive/MyDrive/graduation project/work/data/'+name)
    count = 1
    print(name)
    for x in a:
        #print(x)
        
        img=Image.open('drive/MyDrive/graduation project/work/data/'+name+'/'+x)
        dst=img.transpose(Image.FLIP_LEFT_RIGHT)#左右互换
        newname='drive/MyDrive/graduation project/work/enhance data/'+name+'/'+'image_enhance1_'+str(count)+'.jpg'
        dst=dst.convert('RGB')
        dst.save(newname)
        count += 1
```

```python

# 上下翻转
for name in flowe_classes:
    a = os.listdir('drive/MyDrive/graduation project/work/data/'+name)
    count = 1
    print(name)
    for x in a:
        #print(x)
        
        img=Image.open('drive/MyDrive/graduation project/work/data/'+name+'/'+x)
        dst=img.transpose(Image.FLIP_TOP_BOTTOM)#左右互换
        newname='drive/MyDrive/graduation project/work/enhance data/'+name+'/'+'image_enhance2_'+str(count)+'.jpg'
        dst=dst.convert('RGB')
        dst.save(newname)
        count += 1
```

```python
import numpy as np
```

```python
# 椒盐噪声
for name in flowe_classes:
    a=os.listdir('drive/MyDrive/graduation project/work/data/'+name) 
    count=1
    print(name)
    for x in a:
        oldname='drive/MyDrive/graduation project/work/data/'+name+'/'+x
        img=np.array(Image.open(oldname)) 
        #随机生成5000个椒盐
        rows,cols,dims=img.shape
        for i in range(5000):
            x=np.random.randint(0,rows)
            y=np.random.randint(0,cols)
            if np.random.random()<0.5:
              img[x,y,:]=0 
            else:
              img[x,y,:]=255
        img.flags.writeable = True  # 将数组改为读写模式
        dst=Image.fromarray(np.uint8(img))
        newname='drive/MyDrive/graduation project/work/enhance data/'+name+'/'+'image_enhance3_'+str(count)+'.jpg'
        dst=dst.convert('RGB')
        dst.save(newname)
        count+=1
```
