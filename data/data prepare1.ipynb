```
import os
from imutils import paths
import shutil
```

```
from google.colab import drive
drive.mount('/content/drive/')
```

```
base_path = 'drive/MyDrive/graduation project/'
```

```
for i in range(1, 18):
    i = str(i)
    os.makedirs(base_path + 'data/' + i)
```

```
dataset_dir = 'drive/MyDrive/graduation project/17flowers/jpg'
```

```
picture_list = list(paths.list_images(dataset_dir))
picture_list.sort()
print(picture_list)
```

```
pic_num = 0  # 用来计数1360，便于接下来遍历存储图片

for i in range(1, 18):
    value = 0
    file_path = base_path + 'data/' + str(i)
    while value < 80:
        shutil.copy(picture_list[pic_num], file_path)  # 这个是复制语句，将picture_list的图片复制到file文件夹里
        
        pic_num += 1
        value += 1
```
